# Plotting

## Setup

```julia
try
    using Plots
catch ArgumentError
    using Pkg
    Pkg.add("Plots")
    using Plots
end
```

## Plot

```julia
x = 1:0.1:10Ï€ # type \pi <tab> to get pi
```

### Default Backend

```julia
gr() # not required if no other backend has been selected before
```

### Interactive Plotting (no explicit plot object)

```julia
plot(x, sin.(x), label="sin(x)")
```

```julia
plot!(x, cos.(x), label="cos(x)")
```

### Use Explicit Plot Object

```julia
plt = plot()
```

```julia
plot!(plt, x, sin.(x), label="sin(x)")
```

```julia
plot!(plt, x, sin.(x).^2, label="sin^2(x)")
```

### Plotly Backend

For interactive plots

```julia
plotly()
```

```julia
plt2 = plot()
plot!(plt2, x, cos.(x), label="cos(x)")
plot!(plt2, x, cos.(x).^2, label="cos^2(x)")
```
