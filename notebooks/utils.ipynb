## Inspection

### Show Type Hierarchy

```julia
# source: https://discourse.julialang.org/t/planning-large-projects/28562/11
function showtypetree(T::Type, level=0)
    println("\t" ^ level, T, " ($(parentmodule(T)))")
    for t in subtypes(T)
        showtypetree(t, level+1)
    end
end
```

```julia
showtypetree(Number)
```
